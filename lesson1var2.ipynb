```python
# подготовка к Град. спуску

# начальн. значение
xn = 0 
yn = f(xn)

# заводим словарь, где будем хранить все найденные значения функции
Y = {xn: yn}

# шаг ГС (произвольный подибраем)
step = 0.1

# по формуле градиентного спуска получаем все значения x y
for _ in range(25):
    xn = xn - step*derivative(f, xn)
    yn = f(xn)
    Y[xn] = yn

Y
```
